```python
import json
import random
from collections import defaultdict
```

```python
input_file = '/Users/arvin/arvinDoc/PoliTo/2025-2026-1/dnlp/TISER/data/TISER_test.json'

def sample_json(input_path, sample_size=500):
    dataset_groups = defaultdict(list)
    with open(input_path, 'r', encoding='utf-8') as f:
        for line_number, line in enumerate(f, 1):
            line = line.strip()
            if not line:
                continue
            try:
                item = json.loads(line)
                name = item.get('dataset_name')
                if name:
                    dataset_groups[name].append(item)
            except json.JSONDecodeError as e:
                print(e)

    for name, all_items in dataset_groups.items():
        actual_sample_size = min(len(all_items), sample_size)

        random.seed(114)
        sampled_items = random.sample(all_items, actual_sample_size)
        safe_name = "".join([c for c in name if c.isalnum() or c in (' ', '_')]).rstrip()
        output_filename = f"/Users/arvin/arvinDoc/PoliTo/2025-2026-1/dnlp/TISER/data/{safe_name}_random_500.json"

        with open(output_filename, 'w', encoding='utf-8') as f:
            json.dump(sampled_items, f, ensure_ascii=False, indent=4)

        print(f"Dataset: {name} | Total: {len(all_items)} | Sample: {len(sampled_items)} -> Saved to: {output_filename}")
```

```python
sample_json(input_file)
```
